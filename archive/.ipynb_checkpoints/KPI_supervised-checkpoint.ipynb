### Visualization

```python
df = train[train["KPI ID"] == kpi_ids[0]]
df["timestamp"] = df["timestamp"].apply(lambda x: datetime.utcfromtimestamp(int(x)).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
df.head()
```

```python
anomaly = df[df["label"] == 1]
anomaly.head()
```

```python
plt.figure(figsize=(40,10))
anomaly = df[df["label"] == 1].index
plt.scatter(anomaly, df.iloc[anomaly, 1], marker='o', s=40, c="red")
plt.plot(np.arange(len(df)), df["value"], c="lightblue")
plt.show()
```
